## 2. Apply a filter and sort Revenue in descending order

Filter-like operations use Boolean indexing; sorting uses `sort_values()`.

```python
sorted_by_revenue = df.sort_values('Revenue', ascending=False).reset_index(drop=True)
sorted_by_revenue
```

## 3. Calculate Total Revenue

Use the `sum()` method.

```python
total_revenue = df['Revenue'].sum()
total_revenue
```

## 4. Calculate Average Revenue

Use the `mean()` method.

```python
average_revenue = df['Revenue'].mean()
average_revenue
```

## 5. Count total records

`len(df)` counts rows, including rows with missing values in some columns.

```python
total_records = len(df)
total_records
```

## 6. Count numeric Revenue cells

`count()` counts non-missing values in the Revenue column.

```python
numeric_revenue_cells = df['Revenue'].count()
numeric_revenue_cells
```

## 7. COUNTIF: Orders from North region

Create a Boolean condition and sum `True` values.

```python
north_orders = df['Region'].eq('North').sum()
north_orders
```

## 8. SUMIF: Total Revenue for Electronics

Filter the category and sum Revenue.

```python
electronics_revenue = df.loc[df['Category'].eq('Electronics'), 'Revenue'].sum()
electronics_revenue
```

## 9. AVERAGEIF: Average Units sold by Asha

Filter the salesperson and calculate the mean Units.

```python
asha_average_units = df.loc[df['Salesperson'].eq('Asha'), 'Units'].mean()
asha_average_units
```

## 10. Highlight Revenue greater than 1,000

Use a pandas `Styler` to apply conditional formatting.

```python
def highlight_high_revenue(value):
    return 'background-color: lightgreen; font-weight: bold' if value > 1000 else ''

df.style.map(highlight_high_revenue, subset=['Revenue']).format({'UnitPrice': '{:,.2f}', 'Revenue': '{:,.2f}'})
```

## 11. Apply a colour scale to Revenue

Use `background_gradient()` on the Revenue column.

```python
df.style.background_gradient(subset=['Revenue'], cmap='RdYlGn').format({'UnitPrice': '{:,.2f}', 'Revenue': '{:,.2f}'})
```

## 12. Identify duplicate OrderID values

Use `duplicated(keep=False)` to show every occurrence of a repeated ID.

```python
duplicate_orders = df[df['OrderID'].duplicated(keep=False)]
duplicate_orders
```

## 13. Create a lookup table and perform a VLOOKUP-style lookup

In pandas, `map()` or `merge()` replaces VLOOKUP. First create a unique Product–Category lookup table.

```python
product_lookup = df[['Product', 'Category']].drop_duplicates().sort_values('Product').reset_index(drop=True)
product_to_category = product_lookup.set_index('Product')['Category']
df_vlookup = df[['OrderID', 'Product']].copy()
df_vlookup['Category_from_lookup'] = df_vlookup['Product'].map(product_to_category)
product_lookup, df_vlookup.head(10)
```

## 14. Repeat using an XLOOKUP-style approach

A left `merge()` is the closest general equivalent of XLOOKUP.

```python
df_xlookup = df[['OrderID', 'Product']].merge(product_lookup, on='Product', how='left', validate='many_to_one')
df_xlookup.head(10)
```

## 15. QUERY pivot: Sum of Revenue by Region

Use `groupby()` or `pivot_table()`.

```python
revenue_by_region = df.groupby('Region', as_index=False)['Revenue'].sum().sort_values('Revenue', ascending=False)
revenue_pivot = df.pivot_table(index='Region', values='Revenue', aggfunc='sum').sort_values('Revenue', ascending=False)
revenue_by_region, revenue_pivot
```

## 16. SORT formula: Display data by Units descending

Use `sort_values()`.

```python
units_descending = df.sort_values(['Units', 'Revenue'], ascending=[False, False]).reset_index(drop=True)
units_descending
```

## 17. Find highest and lowest Revenue

Use `max()` and `min()`; `idxmax()` and `idxmin()` identify the corresponding rows.

```python
highest_revenue = df['Revenue'].max()
lowest_revenue = df['Revenue'].min()
highest_row = df.loc[df['Revenue'].idxmax()]
lowest_row = df.loc[df['Revenue'].idxmin()]
{'Highest Revenue': highest_revenue, 'Lowest Revenue': lowest_revenue}, highest_row, lowest_row
```

## 18. Calculate each row's percentage contribution to total Revenue

Divide each row's Revenue by total Revenue and multiply by 100.

```python
df_percentage = df.copy()
df_percentage['Revenue_Percentage'] = (df_percentage['Revenue'] / df_percentage['Revenue'].sum() * 100).round(2)
df_percentage[['OrderID', 'Revenue', 'Revenue_Percentage']]
```

## 19. Use IF logic to label Revenue as High or Low

Use `np.where()` for vectorised IF logic.

```python
df_labelled = df.copy()
df_labelled['Revenue_Level'] = np.where(df_labelled['Revenue'] >= 1000, 'High', 'Low')
df_labelled[['OrderID', 'Revenue', 'Revenue_Level']]
```

## 20. Protect formula cells — Python equivalent

A DataFrame has no editable formula cells. Protect the original data by keeping an untouched deep copy and creating derived columns in a separate result DataFrame.

```python
df_original = df.copy(deep=True)
df_results = df_original.assign(
    Revenue_Percentage=lambda x: (x['Revenue'] / x['Revenue'].sum() * 100).round(2),
    Revenue_Level=lambda x: np.where(x['Revenue'] >= 1000, 'High', 'Low')
)
original_unchanged = df_original.equals(df)
original_unchanged, df_results.head()
```
