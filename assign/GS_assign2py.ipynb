## 10. Highlight Revenue greater than 1,000

Use a pandas `Styler` to apply conditional formatting.

```python
def highlight_high_revenue(value):
    return 'background-color: lightgreen; font-weight: bold' if value > 1000 else ''

df.style.map(highlight_high_revenue, subset=['Revenue']).format({'UnitPrice': '{:,.2f}', 'Revenue': '{:,.2f}'})
```

## 11. Apply a colour scale to Revenue

## 13. Create a lookup table and perform a VLOOKUP-style lookup

In pandas, `map()` or `merge()` replaces VLOOKUP. First create a unique Product–Category lookup table.

```python
products = df['Product'].unique()
products
```

```python
df[df['Product'].isin(products)][['Product','Category']].drop_duplicates()
```

```python
#selected products
products1 = ['Chair','Laptop']
df[df['Product'].isin(products1)]
```

## 14. Repeat using an XLOOKUP-style approach

A left `merge()` is the closest general equivalent of XLOOKUP.

```python
products_df = pd.DataFrame({'Product': products})
products_df
```

```python
df[['OrderID', 'Product','Category','Salesperson']].merge(products_df, on='Product', how='left', validate='many_to_one')
```

## 15. QUERY pivot: Sum of Revenue by Region

Use `groupby()` or `pivot_table()`.

```python
# Groupby - recommended
df.groupby('Region', as_index=False)['Revenue'].sum().sort_values('Revenue', ascending=False)
```

```python
# Pivot -> aggfunc
df.pivot_table(index='Region', values='Revenue', aggfunc='sum').sort_values('Revenue', ascending=False)
```

## 18. Calculate each row's percentage contribution to total Revenue

Divide each row's Revenue by total Revenue and multiply by 100.

```python
df.info()
```

```python
df['Revenue_Perc'] = round(100 * df['Revenue'] / df['Revenue'].sum(), 2)
df.head()
```

```python
## method 2: take sum of all row values in a column, divide each column value by Total Revenue
df['Total_Revenue'] = df['Revenue'].sum()
df['Revenue_Perc2'] = round(100 * df['Revenue'] / df['Total_Revenue'], 2)
df.head()
```

## 19. Use IF logic to label Revenue as High or Low

Use `np.where()` for vectorised IF logic.

```python
df['Revenue_Level'] = np.where(df['Revenue'] >= 1000, 'High', 'Low')
```

```python
df[['OrderID', 'Revenue', 'Revenue_Level']].assign(compValue = 1000, Diff= lambda d:d['Revenue']- d['compValue']).head()
```

```python
df[['OrderID', 'Revenue', 'Revenue_Level']].assign(compValue = 1000, Diff= lambda d:d['Revenue']- d['compValue']).style.apply( lambda s: ['background-color: lightgreen' if v == 'High'
else 'background-color: salmon' if v == 'Low'   else '' for v in s],   subset=['Revenue_Level'])
```

## 20. Protect formula cells — Python equivalent

- A DataFrame has no editable formula cells. Protect the original data by keeping an untouched deep copy and creating derived columns in a separate result DataFrame.
- A deep copy of a DataFrame creates a completely independent copy of the data and index. Changes made to the copied DataFrame do not affect the original DataFrame.

```python
df_original = df.copy(deep=True)
df_original.head()
```
