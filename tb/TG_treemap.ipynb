## Python - Tree Map

```python
# Data
import numpy as np
import pandas as pd
import datetime as dt
import plotly.express as px
url1 = 'https://raw.githubusercontent.com/dupadhyaya/piit/refs/heads/main/data/superstore_orders.csv'
print(url1)
sales = pd.read_csv(url1)
sales.shape
```

```python
tree_data = ( sales.groupby(['Category', 'Sub-Category'], as_index=False)\
    .agg(Sales=('Sales', 'sum')))
tree_data.head()
```

```python
fig = px.treemap(tree_data, path=['Category', 'Sub-Category'], values='Sales',    color='Sales')
fig.show();
```
