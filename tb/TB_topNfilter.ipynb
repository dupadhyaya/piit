# Tableau - Top N Filter

## Top N Filter
- Parameter : N =1
- Calculation
- Filter

-  In Tableau, you can use a Parameter to allow users to dynamically choose the Top X States by Sales in the Superstore dataset.

## Step 1: Create a Parameter
1. Right-click in the Data pane → Create Parameter
2. Configure:
    * Name: Top N States
    * Data Type: Integer
    * Current Value: 10
    * Allowable Values: Range
        * Minimum: 1
        * Maximum: 50
        * Step Size: 1
3. Click OK
4. Right-click the parameter → Show Parameter

⸻

## Step 2: Create a Rank Calculation
- Create a calculated field named State Rank:
- RANK(SUM([Sales]), 'desc')

⸻

## Step 3: Create a Filter Calculation
- Create another calculated field named Top N Filter:
- RANK(SUM([Sales]), 'desc') <= [Top N States]
or

IF RANK(SUM([Sales]), 'desc') <= [Top N States]
THEN "Keep"
END

⸻

## Step 4: Build the View

1. Drag State to Rows
2. Drag Sales to Columns
3. Sort descending by Sales

⸻

## Step 5: Apply the Filter

- Drag Top N Filter to Filters shelf.

* If using Boolean calculation:
    * Select True
* If using “Keep” calculation:
    * Select Keep

⸻

## Step 6: Set Table Calculation
- Since RANK() is a table calculation:

1. Right-click the Top N Filter pill
2. Choose Edit Table Calculation
3. Compute Using → State

This ensures ranking is calculated across all states.

⸻

## Alternative (Recommended): Using Tableau’s Built-in Top Filter
- Instead of Rank calculations:

1. Drag State to Filters
2. Go to Top tab
3. Select:
    * By Field
    * Top → Parameter: Top N States
    * Sum(Sales)

## Example:
- Top [Top N States] by Sum(Sales)
- This approach is simpler and performs better on large datasets.

⸻

## Result
- The user can change the parameter value:

Parameter Value	Output
5	Top 5 States by Sales
10	Top 10 States by Sales
20	Top 20 States by Sales

and the visualization updates automatically.

- Parameter - TopNStates
- Calculated Field - RankState, TopNFilter
- Add to. Filter : TopNFilter - True
- Show Parameter : Change Values

## Python - Top X States

```python
# Data
import numpy as np
import pandas as pd
import datetime as dt
url1 = 'https://raw.githubusercontent.com/dupadhyaya/piit/refs/heads/main/data/superstore_orders.csv'
print(url1)
sales = pd.read_csv(url1)
sales.shape
```

```python
sales=sales.rename(columns = {'State/Province':'State', 'Country/Region':'Region'})
```

```python
sales.columns
```

```python
try:
    x = int(input('Top N state Count by Sum of Sales: '))
    print(f"You entered {x}")
except ValueError:
    print("Please enter a valid integer.")
```

```python
x
```

```python
sales.groupby(['State'])['Sales'].sum().nlargest(x)
```

```python
top_n_states = ( sales.groupby('State', as_index=False)['Sales']
    .sum().sort_values('Sales', ascending=False).head(x))
top_n_states.insert(0, 'Rank', range(1, len(top_n_states) + 1))
top_n_states
```
