## Python Groups and Sets
### Groups

```python
# Data
import numpy as np
import pandas as pd
import datetime as dt
import plotly.express as px
url1 = 'https://raw.githubusercontent.com/dupadhyaya/piit/refs/heads/main/data/superstore_orders.csv'
print(url1)
sales = pd.read_csv(url1)
sales.shape
```

```python
sales.columns
```

```python
sales=sales.rename(columns = {'State/Province':'State', 'Country/Region':'Region'})
#old->new
sales.columns
```

```python
# Create another column : stateGroup1 by first character of the  state
def state_group(state):
    first = state[0].upper()
    if 'A' <= first <= 'G':
        return 'A-G'
    elif 'I' <= first <= 'M':
        return 'I-M'
    elif 'N' <= first <= 'Z':
        return 'N-Z'
    return 'Other'
```

```python
sales['State_Group1'] = sales['State'].apply(state_group)
sales[['State','State_Group1']]
```

```python
sales[['State','State_Group1']].groupby('State_Group1').agg('count')
```

###  Sets

```python
sales['Profit'].dtypes
```

```python
#### M1
sales['Profit_Category1'] = np.select(  [  
    sales['Profit'] < 0, sales['Profit'].between(0, 100),
    sales['Profit'].between(100, 500), sales['Profit'] > 500 ], 
    ['Loss','Low Profit','Medium Profit','High Profit' ], default='Unknown')
sales[['Profit', 'Profit_Category1']].head()
```

```python
### M2
sales['Profit_Category2'] = pd.qcut( sales['Profit'],  q=4, \
  labels=['Loss/Very Low', 'Low Profit', 'Medium Profit', 'High Profit'])
sales[['Profit', 'Profit_Category1', 'Profit_Category2']].head()
```

### Sets - Top 3 States 
- Tableau
- 1. Right-click State → Create → Set
- 2. Top tab
- 3. Select Top 5 by Sum(Sales)
- 4. Use the set field in the view (IN = True, OUT = False).

```python
# Get top 5 states by total sales
top_5_states = (sales.groupby('State')['Sales'].sum()
    .nlargest(5).index)
top_5_states
```

```python
sales['Top_5_State'] = sales['State'].isin(top_5_states)
```

```python
sales[['State', 'Sales', 'Top_5_State']].head()
```

```python
list(top_5_states)
```

```python
sales[['State', 'Sales', 'Top_5_State']].groupby(['Top_5_State','State'])['Sales']. \
    sum().sort_index(ascending=False)
```

```python
sales[['State', 'Sales', 'Top_5_State']].groupby(['Top_5_State','State'])['Sales']. \
    sum().sort_values(ascending=False)
```
