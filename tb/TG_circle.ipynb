# Tableau - Graph - Circle Chart and Side By Side Circle
- 

- A Circle Chart displays data as individual points, perfect for spotting distributions or comparing measures across categories.
- A Side-by-Side Circle View expands on this by grouping multiple marks side-by-side, making it easy to compare dimensions (e.g., product subcategories) within a primary category (e.g., customer segments).
## How to Create a Standard Circle View
- Drag Dimensions: Drop a Dimension (e.g., Category) onto the Columns shelf.
- Drag Measures: Drop a Measure (e.g., Sales) onto the Rows shelf.Change Mark Type: In the Marks card, change the dropdown from Automatic to Circle.
- Customize: Drag another dimension to Color to differentiate groups or a measure to Size to highlight values.
## How to Create a Side-by-Side Circle View
- Add Primary Dimension: Drop your main Dimension (e.g., Segment) to the Columns shelf.
- Add Secondary Dimension: Drag a second Dimension (e.g., Sub-Category) and drop it to the right of the first on the Columns shelf.
- Add Measure: Drop your Measure (e.g., Profit) onto the Rows shelf.Select Mark Type: Change the Marks card to Circle.
