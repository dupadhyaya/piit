## Pandas Plot Functions
- DataFrame.plot.hist(by=None, bins=10, **kwargs)
- using kind='hist' within the plot() method allows you to create histograms to visualize the distribution of numerical data . Unlike a bar chart which plots raw values, kind='hist' automatically bins the data and counts frequencies
- boxplot
    - https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.boxplot.html 

```python
 # 1. Histogram
df['mpg'].plot(kind='hist', title='Distribution of MPG')
plt.show();
```

```python
#df['mpg'].plot(kind='hist', bins=5, title='Distribution of MPG')
df['mpg'].plot(kind='hist', bins=[0,10,20,30,35,40], title='Distribution of MPG : Count in Class  Intevals', edgecolor='black', xlabel='MPG', ylabel='Count')
plt.show();
```

```python
# 2. Scatter Plot
df.plot(kind='scatter', x='wt', y='mpg', title='Wt vs MPG')
plt.show();
```

```python
# 3. Boxplot
df.boxplot(column='mpg', by='cyl', patch_artist=True, color=dict(boxes='brown', whiskers='orange', medians='r', caps='g'))
plt.suptitle('Box Plot : cyl vs mpg')
plt.show();
```

## Line
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.line.html
- DataFrame.plot.line(x=None, y=None, color=None, **kwargs)[source]
- This Pandas code sorts a DataFrame (df) by the 'wt' (weight) column in ascending order, then creates a line plot of 'wt' (x-axis) vs 'mpg' (y-axis) titled "Weight vs MPG". This is ideal for visualizing trends between continuous variables, where the sorted x-axis ensures the line connects points in order, creating a smooth line plot

```python
df[['wt','mpg']].sort_values('wt')
```

```python
# 4. Line Plot
df.sort_values('wt').plot(x='wt', y='mpg', kind='line', title='Weight vs MPG')
#df.plot(x='wt', y='mpg', kind='line', title='Weight vs MPG')
plt.show();
```
