# Label Data in Graphs
- Matplotlib doesn’t have one universal labeling function for every chart type. However, there are a few simple patterns that cover almost everything. (like ggplot)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
df = pd.DataFrame({    "Campus": ["Noida", "Jaipur", "Lucknow", "Mumbai"],
    "Admissions": [1200, 850, 650, 780],
    "Revenue": [145.36, 1092.78, 71.555, 88.4],
    "Applications": [6200, 4100, 3300, 3900]})
df
```

```python
fig, ax = plt.subplots(figsize=(8,5))
bars = ax.bar(df["Campus"], df["Admissions"])
ax.bar_label( bars, fmt="%.1f", padding=3) # eq geom_text(aes(label=Admissions)), %.0f
ax.set_title("Campus-wise Admissions")
ax.set_ylabel("Admissions")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
bars = ax.bar(df["Campus"], df["Revenue"])
ax.bar_label( bars, label = [f"{x:,.4f}" for x in df['Revenue'] ], padding=5) 
#ax.bar_label( bars, label = [f"{x:,}" for x in df['Admissions'] ], padding=3) 
ax.set_title("Campus-wise Revenue")
ax.set_ylabel("Revenue")
plt.tight_layout()
plt.show()
```

```python
## Function
def add_labels(ax, fmt="{:,.0f}"):
    for container in ax.containers:
        ax.bar_label(  container,  labels=[fmt.format(v) for v in container.datavalues], padding=3)
#The padding argument in ax.bar_label() controls the distance between the bar and its label, 
#measured in points (a typography unit; 72 points = 1 inch). It is not measured in your data units
#* % → start of a format specification.
#* f → format the value as a floating-point number.
#* .0 → display 0 digits after the decimal point.
```

```python
# bars = ax.bar(df["Campus"], df["Admissions"]) #add_labels(ax)
fig, ax = plt.subplots(figsize=(8,5))
bars = ax.bar(df["Campus"], df["Revenue"])
add_labels(ax)
ax.set_title("Campus-wise Revenue")
ax.set_ylabel("Revenue")
plt.tight_layout()
plt.show()
```
