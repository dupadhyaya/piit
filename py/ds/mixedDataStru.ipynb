## List 
- An ordered and mutable collection that allows duplicate elements. Defined using square brackets [ ], lists are highly versatile for storing heterogeneous data

```python
myList1 = [1, 'a', 'Dhiraj', True, 101, 100.256 ]
```

```python
myList1
```

```python
type(myList1)
```

```python
myList1[1]
```

## Tuple
- An ordered but immutable collection. Once created by ( ), elements cannot be modified, making them ideal for fixed data like coordinates or RGB values

```python
myList2 = (1, 'a', 'Dhiraj', True, 101, 100.256 , 'Mandal')
```

```python
myList2
```

```python
type(myList2)
```

```python
myList2[1] = 'b'
```

```python
myList2 = (1,'Rajia', 'B', True)
myList2
```
