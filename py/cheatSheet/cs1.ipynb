## Subset Rows and Columns
- Use df.loc[] and df.iloc[] to select only rows, only columns or both.
- Use df.at[] and df.iat[] to access a single value by row and column.
- First index selects rows, second index columns.

```python
mtcars.iloc[10:15]
```

```python
mtcars.iloc[:,[1,2,5]].head()
```

```python
mtcars.loc[:, 'mpg':'wt'].head()  #names of columns
```

```python
mtcars.loc[:, ['mpg','wt','am']].head() 
```

```python
mtcars.loc[mtcars['mpg'] > 30, ['mpg','wt','am']].head() 
```

```python
mtcars.iat[1,2]  # by index
```

```python
mtcars.at['Fiat 128','am']  #by  label
```
