## Use of brackets in Python
- Round ( ) ; Square [  ] ; Curly { }
- This is a very important foundational concept—and often confused by beginners.
- The best way to learn it is by linking each bracket to its purpose + data type + usage.
- “Each bracket represents a different way to organize data”

🔵 Types of Brackets in Python & Their Meaning

## 🟢 1. Round Brackets ( )
- 📌 Used for:
    - Functions
    - Grouping expressions
    - Tuples
- 🧠 Key Idea:
    - ( ) = do something / group things

```python
# Function call
print("Hello")
# Math grouping
result = (2 + 3) * 5
print(result)
# Tuple
t = (1, 2, 3)
print(t)
```

## 🟦 2. Square Brackets [ ]
- 📌 Used for:
    * Lists
    * Indexing
    * Slicing
- 🧠 Key Idea:
    - [ ] = store multiple items OR access them

```python
#✅ Examples
# List
numbers = [10, 20, 30]
# Indexing
print(numbers[0])   # 10
# Slicing
print(numbers[1:])  # [20, 30]
```

## 🟨 3. Curly Brackets { }
- 📌 Used for:
    - Dictionaries
    - Sets
- 🧠 Key Idea:
    - { } = mapping OR unique collection

```python
##✅ Examples
# Dictionary (key-value pairs)
student = {"name": "EDA", "age": 21}
print(student["name"])
# Set (unique values)
s = {1, 2, 3, 3}
print(s)   # {1, 2, 3}
print(type(s))
```

🔁 Quick Comparison Table
Bracket	Name	Used For	Example
( )	Round	Functions, tuples	(1,2)
[ ]	Square	Lists, indexing	[1,2,3]
{ }	Curly	Dict, set	{"a":1}

⚠️ Common Beginner Confusions
- ❌ Mistake 1: Mixing list & tuple
    - a = (1, 2, 3)   # tuple (immutable)
    - b = [1, 2, 3]   # list (mutable)
⸻
- ❌ Mistake 2: Dictionary vs Set
    - a = {}        # dictionary
    - b = {1, 2}    # set
⸻
- ❌ Mistake 3: Forgetting brackets in function
    - print      # ❌ wrong
    - print()    # ✅ correct
⸻
- 🎯 Learning Trick (Very Effective) analogy:
    - ( ) → Action / execution
    - [  ] → Collection / storage/ indexing
    - { } → Mapping / relationship
