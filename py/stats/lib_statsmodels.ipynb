# Library - statsmodels
- Statsmodels is a Python library designed for statistical computations, offering a wide range of functionalities for statistical modeling, testing, and data exploration. It provides a comprehensive set of tools for various statistical tasks, complementing numerical libraries like NumPy and SciPy.
- https://www.statsmodels.org/stable/index.html
- https://www.statsmodels.org/stable/user-guide.html
- https://www.statsmodels.org/stable/examples/index.html

```python
import statsmodels.api as sm
import pandas
from patsy import dmatrices
```

```python
df = sm.datasets.get_rdataset("Guerry", "HistData").data
df.head()
```

```python
y, X = dmatrices('Lottery ~ Literacy + Wealth + Region', data=df, return_type='dataframe')
```

```python
y
```

```python
X
```

```python
mod = sm.OLS(y, X) 
```

```python
res = mod.fit() 
```

```python
print(res.summary())
```

```python
res.params
```

```python
res.rsquared
```

```python
sm.stats.linear_rainbow(res)
```

## ANOVA
- https://www.statsmodels.org/stable/anova.html

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
moore = sm.datasets.get_rdataset("Moore", "carData", cache=True) # load data
data = moore.data
data.head()
```

```python
data = data.rename(columns={"partner.status": "partner_status"}) # make name pythonic
```

```python
moore_lm = ols('conformity ~ C(fcategory, Sum)*C(partner_status, Sum)', data=data).fit()
```

```python
table = sm.stats.anova_lm(moore_lm, typ=2) # Type 2 ANOVA DataFrame
```
