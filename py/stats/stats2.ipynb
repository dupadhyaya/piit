## Statistics - 2 (Distributions) 
-  core statistics concept that every data analyst must understand.
- Distributions
    - A distribution describes how the values of a variable are spread or dispersed. It tells us:
        - What values occur,
        - How frequently they occur,
        - The shape of the data.
    - In other words, it shows how data behaves — whether it’s centered, skewed, spread out, or has outliers.
- Examples
    - The height of people → roughly normal distribution
    - Number of customer purchases → often Poisson
    - Probability of getting heads → Binomial
- Types
    - Discrete - values are countable figures
        - Uniform (Discrete) - All outcomes equally likely, Rolling a die (1–6)
        - Binomial - Success/failure repeated n times, Flipping coins
        - Poisson - Counts in a fixed time/space, Number of arrivals per minute
    - Continuous - Values can take any number within a range.
        - Uniform (Continuous) - Equal probability within range, Random number between 0–1
        - Normal (Gaussian)- Bell-shaped, symmetric; Human height, test scores
        - Exponential - Time between random events, Wait time between customers
        - Gamma / Beta / Lognormal - Used in reliability, survival, etc., Advanced modeling

### Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=0, scale=1)   # mean=0, std=1
print(x [0:10],y[0:10])
```

```python
plt.plot(x, y)
plt.title("Normal Distribution (mean=0, std=1)")
plt.xlabel("x"); plt.ylabel("Probability Density")
plt.show();
```

## Uniform Distribution

```python
from scipy.stats import uniform

x = np.linspace(0, 1, 1000)
y = uniform.pdf(x, loc=0, scale=1)
print(x[0:10], y[0:10])
```

```python
plt.plot(x, y)
plt.title("Uniform Distribution (0 to 1)")
plt.show();
```

## Binomial

```python
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

plt.bar(x, y)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Successes"); plt.ylabel("Probability")
plt.show()
```

## Poision

```python
from scipy.stats import poisson

x = np.arange(0, 15)
y = poisson.pmf(x, mu=5)

plt.bar(x, y)
plt.title("Poisson Distribution (λ = 5)")
plt.show()
```

## Exponential Distribution

```python
from scipy.stats import expon

x = np.linspace(0, 10, 1000)
y = expon.pdf(x, scale=2)

plt.plot(x, y)
plt.title("Exponential Distribution (scale=2)")
plt.show()
```

## Show multiple distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, poisson

# Create figure
plt.figure(figsize=(10,6))

# Normal Distribution (mean=0, std=1)
x_norm = np.linspace(-4, 4, 1000)
y_norm = norm.pdf(x_norm, loc=0, scale=1)
plt.plot(x_norm, y_norm, label='Normal (μ=0, σ=1)', linewidth=2)

# Uniform Distribution (0 to 1)
x_uni = np.linspace(0, 1, 1000)
y_uni = uniform.pdf(x_uni, loc=0, scale=1)
plt.plot(x_uni, y_uni, label='Uniform (0–1)', linewidth=2)

# Poisson Distribution (λ=5)
x_pois = np.arange(0, 15)
y_pois = poisson.pmf(x_pois, mu=5)
plt.stem(x_pois, y_pois, linefmt='C2-', markerfmt='C2o', basefmt=" ", label='Poisson (λ=5)')

# Styling
plt.title("Comparison of Different Distributions", fontsize=14)
plt.xlabel("x")
plt.ylabel("Probability / Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
