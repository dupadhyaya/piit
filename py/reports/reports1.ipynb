# Report Formats
- output formats

```python
import pandas as pd
from pydataset import data
import matplotlib.pyplot as plt
```

```python
mtcars = data('mtcars')
mtcars.head()
```

## IPython.display (Recommended)

```python
from IPython.display import Markdown, display
display(Markdown("# Sales Report"))
display(Markdown("""
## Summary
- Total Sales: **₹10 Cr**
- Growth: **12%**
- Best Campus: **Noida**
"""))
```

## Pandas to_markdown()

```python
#pip install tabulate
```

```python
df = pd.DataFrame({    "Campus":["A","B","C"],   "Sales":[120,150,180]})
print(df.to_markdown(index=False))
```

```python
print(mtcars.to_markdown(index=False))
```

```python
from IPython.display import Markdown, display
display(Markdown(df.to_markdown(index=False)))
```

```python
display(Markdown(mtcars.to_markdown(index=False)))
```

## Tabulate

```python
from tabulate import tabulate
table = tabulate(df, headers="keys",   tablefmt="github",  showindex=False)
print(table)
```

```python
table2 = tabulate(mtcars, headers="keys",   tablefmt="grid",  showindex=False) ##tablefmt = pipe, github, grid, simple
print(table2)
```

## Combine Markdown + Data

```python
from IPython.display import Markdown, display
display(Markdown(f"""
# Monthly Report
## KPIs
- Total Students: **{5200}**
- Admissions: **{350}**
- Revenue: **₹12.5 Cr**
"""))
```

```python
report = f"""
## Admissions Dashboard
### Highlights
- Total Applications: **{12000}**
- Offers: **4300**
- Enrolments: **2700**
### Top Campus
***New York***
"""
display(Markdown(report))
```

## Markdown Tables + Charts

```python
display(Markdown("## Revenue by Campus"))
display(mtcars)
plt.show()
```

```python
display(Markdown(df.to_markdown(index=False)))
plt.show()
```

## Jinja2 (Professional Reports)

```python
from jinja2 import Template
template = Template("""
# {{ title }}
Total Students: **{{ students }}**
Revenue: **₹{{ revenue }} Cr**
""")
md = template.render(
    title="Admissions Report",
    students=5200,
    revenue=18.2
)
display(Markdown(md))
```

## Summary
* IPython.display.Markdown for headings, commentary, and narrative.
* pandas.to_markdown() (or tabulate) for clean tables.
* Plotly or Matplotlib for charts.
* Jinja2 if you want reusable report templates.
* Quarto if you need to publish polished reports to HTML, PDF, or Word from your notebooks.
