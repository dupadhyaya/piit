# Date and Time in 
- How to get current date & time
- How to format date/time
- How to extract parts (year, month, etc.)
- How to perform basic operations

```python
from datetime import datetime
# datetime is the built-in module to work with date & time
```

```python
# Current Date & Time
now = datetime.now()

print("Current Date & Time:", now)
```

```python
# Extracting Components
print("Current Date & Time:", now)
print("\t Year:", now.year, end ='\t')
print("\t Month:", now.month, end ='\t')
print("\t Day:", now.day, end ='\t')
print("\t Hour:", now.hour, end ='\t')
print("\t Minute:", now.minute, end ='\t')
print("\t Second:", now.second, end ='\t')
```

```python
# Formatting Date
formatted = now.strftime("%d-%m-%Y")
print("Formatted Date:", formatted)
```

```python
formatted2 = now.strftime("%d-%b-%Y")
print("Formatted Date:", formatted2)
```

## Common Formats
- %d - day (30);
- Month : %m (6), %b (Jun), %B (June)
- Year : %y (26), %Y (2026)
- Hour : %H
- Minute : %M
- Second : %S

```python
# String to Date
date_str = "26-04-2026"
date_obj = datetime.strptime(date_str, "%d-%m-%Y")
print(date_obj)
```

## Date Arithmetic

```python
from datetime import timedelta
today = datetime.now()
future = today + timedelta(days=7)
past = today - timedelta(days=3)
print("After 7 days:", future)
print("3 days ago:", past)
```

```python
# Difference
date1 = datetime(2026, 3, 15)
date2 = datetime(2026, 4, 26)
difference = date2 - date1
print("Difference:", difference.days, "days")
```

```python
# Only Date
from datetime import date
today = date.today()
print("Today:", today)
```

## Functions
- datetime . full date + time
- date. only date
- timedelta. difference between dates
- strftime. date → string
- strptime. string → date
