## Brackets in Python
- [   ] - square : creating lists, selecting items
- ( ) - round : functions, create tuple object
- { } - curly : set
- {key:value} - dictionary
- " " or ' ' - String

# Data Structures in Python
- Ordered" structures preserve the sequence of elements as they were inserted, while "unordered" structures do not guarantee any specific order, [index]
- Mutable : if its internal state or content can be modified after it has been created: Change internal value
- Heterogenity : ability to store different types of data. 1, 2, 'Dhiraj', True, False
- Ordered
    - Lists, Tuple
    - Strings
- Unordered
    - Sets, Frozen set
    - Dictionary

### Set
- an unordered collection of data that is mutable and does not allow any duplicate element

```python
groupA = set(['England','USA', 'Spain', 'India'])
groupB = set(['India','France','Poland','India'])
```

```python
print(groupA, "\n", groupB)
```

```python
groupA.union(groupB)
```
