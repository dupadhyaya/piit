# Interview Ques - 
- (Covers basic queries, filtering, aggregation, and joins)

- What’s the difference between WHERE and HAVING clauses?
- How do you select all columns from a table?
- Write a query to select only unique values from a column.
- How do you sort query results in descending order?
- Write a query to find the top 5 highest sales.
- What’s the difference between INNER JOIN and LEFT JOIN?
- How do you filter rows where a column value is NULL?
- Write a query to count the number of rows in a table.

- How do you group data by a column and get an average?
- How do you rename a column in the SELECT output?
- Write a query to get records where a text column contains “India”.
- How do you use BETWEEN to filter a date range?
- What’s the difference between UNION and UNION ALL?
- Write a query to find duplicate values in a table.
- How do you limit results to the first 10 rows?
- Write a query to calculate the percentage of total sales for each category.
- How do you join more than two tables?
- How do you create an alias for a table?
- What’s the difference between a primary key and a foreign key?
- How do you update a specific record in a table?
