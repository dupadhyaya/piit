## Interview Process

## Stage 1 – Introduction (5 Minutes)

### Objectives

* Build rapport
* Reduce candidate anxiety
* Explain interview flow

### Sample Questions

* Tell us about yourself.
* What interested you in this position?
* Briefly walk us through your professional journey.

### Evaluate

* Confidence
* Communication
* Clarity
* Professionalism

## Stage 2 – Resume & Experience (10 Minutes)

### Objectives

- Validate claims made in the resume.

### Sample Questions

* Tell us about your current role.
* Which project are you most proud of?
* What was the biggest challenge you faced?
* Describe your greatest professional achievement.
* Why are you looking for a change?

### Evaluate

* Depth of experience
* Ownership
* Business impact
* Honesty


## Stage 3 – Technical Assessment (20 Minutes)

- Questions depend on the role but should include:

### Conceptual Knowledge

* Explain an important concept in your field.
* What are the latest developments in your area?

### Practical Knowledge

* How would you solve this problem?
* Which tools have you used?
* What challenges have you faced?

### Hands-on Scenario

Give a practical exercise or ask the candidate to explain how they would approach a real-world task.

Evaluate

* Technical competence
* Logical thinking
* Practical experience
* Depth of understanding

### Stage 4 – Case Study / Problem Solving (10 Minutes)

Present a business or operational problem.

- Example

A company has experienced a decline in customer satisfaction over the past six months. How would you investigate the problem?

### Observe whether the candidate

* Clarifies assumptions
* Asks relevant questions
* Structures the problem
* Prioritizes information
* Suggests practical solutions
* Explains reasoning clearly

### Evaluate

* Analytical thinking
* Decision-making
* Creativity
* Communication

### Stage 5 – Behavioural Interview (10 Minutes)

Ask questions based on past experiences.

Examples

* Tell me about a difficult colleague you worked with.
* Describe a time when you failed.
* How do you handle conflicting priorities?
* Tell us about a time you showed leadership.
* Describe a situation where you had to learn something quickly.
* Tell us about a difficult customer or stakeholder.

### Evaluate

* Emotional intelligence
* Teamwork
* Adaptability
* Leadership
* Integrity

### Stage 6 – Candidate Questions (5 Minutes)

Invite questions such as

* What are the team’s priorities?
* How is performance measured?
* What learning opportunities are available?
* What are the next steps in the recruitment process?

Observe

* Curiosity
* Preparation
* Long-term interest

### General Evaluation Rubric

- Competency	Weight (%)
- Communication Skills	15
- Technical Knowledge	25
- Problem Solving	20
- Practical Experience	15
- Behavioural Competencies	10
- Learning Ability	5
- Team & Culture Fit	5
- Motivation & Professionalism	5
- Total	100

### Interviewer’s Checklist

During the interview, continuously assess whether the candidate:

* Listens carefully before answering.
* Communicates clearly and concisely.
* Supports answers with examples.
* Demonstrates logical thinking.
* Admits when they don’t know something rather than guessing.
* Shows curiosity and willingness to learn.
* Maintains a professional and respectful attitude.
* Responds confidently under pressure.
* Aligns their experience with the role.
* Ends the interview positively by asking thoughtful questions.

This structure is suitable for practice interviews across technical, managerial, and professional roles, and can be adapted by simply changing the technical questions while keeping the overall interview flow the same.

A strong candidate uses the interview not only to answer questions but also to evaluate whether the role and organization are a good fit. Thoughtful questions show preparation, curiosity, and long-term thinking.

Here are examples of good questions, grouped by topic.

1. About the Role

* What are the top priorities for this role during the first three to six months?
* What would success look like after one year?
* What are the biggest challenges the person in this role will face?
* Why is this position open?
* What skills distinguish top performers in this role?

2. About the Team

* Can you tell me about the team I would be working with?
* How is the team structured?
* How do team members typically collaborate?
* What leadership style does the manager follow?
* How are responsibilities divided within the team?

3. About the Organization

* What are the organization’s strategic priorities over the next few years?
* What do you enjoy most about working here?
* How would you describe the company culture?
* What values are most important in the organization?
* How does the organization encourage innovation?

4. Learning and Growth

* What training or learning opportunities are available?
* How is career progression typically managed?
* Are employees encouraged to pursue certifications or higher education?
* What qualities have helped previous employees grow within the organization?

5. Performance Expectations

* How is performance measured for this role?
* How often are performance reviews conducted?
* What key performance indicators (KPIs) would I be expected to achieve?
* What does an outstanding performer usually accomplish?

6. Work Environment

* What tools and technologies does the team use?
* Is the work primarily office-based, hybrid, or remote?
* How are new ideas and suggestions received?
* How are cross-functional teams coordinated?

7. About the Interviewer

* What has your experience been like working here?
* What do you enjoy most about leading your team?
* What characteristics have you observed in your most successful team members?

8. Next Steps

* What are the next stages of the selection process?
* When can I expect to hear back?
* Is there anything in my background that you’d like me to clarify?
* Is there any area where you feel I could provide more information?

Questions Candidates Should Avoid

* Questions whose answers are clearly available on the company’s website (e.g., “What does your company do?”).
* Questions focused only on salary, leave, or perks in the first interview, unless the interviewer raises them.
* Personal questions about the interviewer.
* Questions that suggest a lack of interest, such as “How soon can I take leave?” or “How much overtime is expected?” as the opening question.

The Three Best Questions to End an Interview

If time is limited, these three leave a particularly positive impression:

1. What would success look like for the person selected for this role in the first six months?
2. What qualities differentiate someone who performs well in this position from someone who performs exceptionally well?
3. Is there anything we’ve discussed today that gives you any hesitation about my suitability for the role? I’d appreciate the opportunity to address it.
