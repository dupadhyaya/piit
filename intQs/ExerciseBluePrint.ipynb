## Formal Job Description (JD)
- Job Title: Data Analyst
- Location: Remote / Hybrid
- Department: Business Intelligence & Analytics
- Position Overview
    - We are seeking a data-driven Data Analyst to transform raw data into actionable business insights. In this role, you will handle data collection, cleaning, modeling, and visualization.
    - You will work closely with stakeholders across departments/ clients to optimize operations and drive data-informed decisions.
- Key Responsibilities
    - Extract, clean, and stage large datasets from transactional databases. Develop, maintain, and automate operational dashboards for senior leadership.
    - Perform exploratory data analysis (EDA) to uncover business trends and anomalies.
    - Translate complex data findings into clear, non-technical business recommendations.
- Required Technical Stack & Skills
    - Excel: Mastered use of XLOOKUP, VLOOKUP, Pivot Tables, and Data Modeling (Power Pivot).
    - MySQL: Confident writing complex queries, multi-table JOIN operations, subqueries, and window functions.
    - Python: Intermediate scripting using pandas and numpy for data manipulation, cleaning, and EDA.
    - Tableau: Ability to connect to data sources, build interactive dashboards, create calculated fields, and apply visual design best practices.
- Soft Skills
    - Strong verbal and written communication.
    - Analytical mindset with sharp attention to detail.
    - Ability to explain technical workflows to non-technical stakeholders.

## Step 4: Interview Execution Guidelines
- Each interview round lasts 15-30 minutes total and follows a strict structured format.
- For the Interviewer (15 Minutes Asking)
    - The Icebreaker (2 mins): "Tell me about yourself and your journey into data analytics.
    - "The Technical Deep-Dive (8 mins):
        - Open their GitHub/Tableau Public links.
        - Ask: "Walk me through the logic of this SQL query/Python script."
        - Ask: "Why did you choose this specific chart type for your Tableau dashboard?"
        - Ask: "How did you handle missing or dirty data in your Excel/Python project?"
- Behavioral Questions (5 mins):
    - "Describe a time you got stuck on a technical bug. How did you resolve it?"
- For the Candidate (5 Minutes Asking)The Reverse Interview:
    - The candidate must ask the interviewer 2–3 smart, business-focused questions about the data team, infrastructure, or company goals.
    - Feedback Session (10 Minutes)
    - The interviewer provides immediate constructive feedback.Discuss portfolio presentation, communication clarity, and technical confidence.
    - Switch roles and repeat the exact process for Round 2.

## Part-2 : Interview Prompts
- Interviewers should pick one MySQL and one Python prompt to test the candidate's portfolio and live logic.
### MySQL Technical Prompts
    - Portfolio Drill-Down: "Open your GitHub SQL file. Walk me through your most complex query. Why did you choose a subquery over a CTE (or vice versa) here?
    - "The Aggregation Scenario: "If I have an orders table and a customers table, write me a conceptual query to find the top 5 customers by total spending in 2025. What JOIN type do you use and why?
    - "Performance & Logic: "Explain the difference between a WHERE clause and a HAVING clause. Can you give me an example where swapping them breaks the query?"

### Python Technical Prompts
- Portfolio Drill-Down: "Open your GitHub Python repository. Look at your data cleaning script. Explain how you handled missing values (NaN) or duplicate rows in your dataset."
- Pandas Mechanics: "If you have a DataFrame called df with columns City and Revenue, how would you find the average revenue grouped by each city using pandas?
- "Error & Bug Handling: "Tell me about a time your Python script threw a frustrating error (like a KeyError or ValueError). How did you debug it?"

## Part 3: Interview Evaluation & Grading Rubric
- Interviewers will score candidates on a scale of 1 to 5 for each category.
- 
### Evaluation Categories
- Portfolio Readiness (GitHub, Tableau, LinkedIn)
    - Criteria: Are links clickable in the CV? Is the GitHub repository organized with a clean README? Is the Tableau dashboard interactive and visually polished?
- Technical Articulation (SQL & Python)
    - Criteria: Can the candidate explain their code step-by-step without getting confused? Do they use correct technical terminology (e.g., "aggregations", "dataframes", "left joins")
- Problem-Solving & Data Cleaning Approach
    - Criteria: How do they approach messy data? Do they show a systematic way of thinking when answering hypothetical scenario questions?
- Communication & Cultural Fit
    - Criteria: Do they maintain good eye contact/camera presence? Is their story clear? Did they ask intelligent, strategic questions during the reverse interview?
